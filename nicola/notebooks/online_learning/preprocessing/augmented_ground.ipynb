## Fix Timin issue

```python
df.reset_index(inplace=True)
df.rename(columns={'index': 'Date'}, inplace=True)
```

```python
df['Date'] = df['Date'].astype(str)
df['Timestamp'] = df['Timestamp'].astype(str)
df['Time'] = pd.to_datetime(df['Date'] + ' ' + df['Timestamp'])
print(df['Time'])
```

```python
df.head()
```

```python
df.at[0, 'Tdiff[s]'] = 0
df.at[1, 'Tdiff[s]'] = 0.700
```

```python
df['Time'] = pd.to_datetime(df['Time'], format='%d/%m/%y %H:%M:%S')
df['Cumulative Tdiff [s]'] = df['Tdiff[s]'].cumsum()
initial_timestamp = df['Time'].iloc[0]
df['Time'] = initial_timestamp + pd.to_timedelta(df['Cumulative Tdiff [s]'], unit='s')
df['Time'] = df['Time'].dt.strftime('%Y/%m/%d %H:%M:%S')
df = df.drop(columns=['Tdiff[s]', 'Cumulative Tdiff [s]'])
```

## retrieve the single cycle

```python
tmp = df.iloc[850:2250]
```

```python
tmp['Current(A)'].plot()
```

## Multiply the single cycle n=1000 times

```python
n_times = 1000
dfs = []
for i in range(n_times):
    df_shifted = tmp.copy()
    df_shifted['Time'] = pd.to_datetime(df_shifted['Time'])
    df_shifted['Time'] += pd.Timedelta(seconds=i * len(tmp))
    dfs.append(df_shifted)
df_repeated = pd.concat(dfs, ignore_index=True)   
```

```python
df_repeated.head()
```

```python
df_repeated.rename(columns={'TempeAmb':'Tamb'}, inplace=True)
```

```python
df_repeated.to_csv('ModuleGround.csv', index=False)
```
