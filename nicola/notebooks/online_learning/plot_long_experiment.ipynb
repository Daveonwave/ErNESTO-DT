```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_to_read = ['../../data/output/driven/testprofile_for_ol_trial_due/long_experiment/2024_08_07-17_16_11/dataset_0.csv',
               '../../data/output/driven/testprofile_for_ol_trial_due/long_experiment/2024_08_07-17_16_11/dataset_1.csv',
               '../../data/output/driven/testprofile_for_ol_trial_due/long_experiment/2024_08_07-17_16_11/dataset_2.csv',
               '../../data/output/driven/testprofile_for_ol_trial_due/long_experiment/2024_08_07-17_16_11/dataset_3.csv',
               '../../data/output/driven/testprofile_for_ol_trial_due/long_experiment/2024_08_07-17_16_11/dataset_4.csv']
df_0 = pd.read_csv(csv_to_read[0])
df_1 = pd.read_csv(csv_to_read[1])
df_2 = pd.read_csv(csv_to_read[2])
df_3 = pd.read_csv(csv_to_read[3])
df_4 = pd.read_csv(csv_to_read[4])
```

```python
df = pd.concat([df_0,df_1,df_2,df_3,df_4], ignore_index=True)
```

```python
df.describe()
```

```python
df.to_csv('long_dataset.csv', index=False)
```
