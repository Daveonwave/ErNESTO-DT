```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = 'datasets_after_square_current/'
dest = '../../data/ground/ground_'
```

```python
for region in ['HH', 'HL', 'LH', 'LL']:
    ground_file = f'ground_{region}.csv'
    df = pd.read_csv(path + ground_file)
    df = df[['time', 'current', 'voltage', 't_amb', 'temperature']]
    df['current'] *= -1
    df = df.rename(columns={'time': 'Time', 'current': 'Current [A]', 'voltage': 'Voltage [V]', 't_amb': 'Ambient Temperature [K]', 'temperature': 'Temperature [K]'})
    df = df.iloc[1:]
    df.to_csv(dest + region + '.csv', index=False)
```
