### Creation of different ambient temperature signals

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
temps = np.arange(0, 42.5, 2.5)
```

```python
signals = {temp: np.random.normal(loc=temp, scale=.1, size=1000000) for temp in temps}
```

```python
signals
```

```python
plt.plot(range(40000), signals[10.0][:40000])
```

```python
df_signals = pd.DataFrame.from_dict(signals)
df_signals
```

```python
for temp in temps:
    df_signals[temp].plot()
```
