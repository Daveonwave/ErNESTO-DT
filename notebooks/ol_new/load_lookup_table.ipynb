```python
import pandas as pd
```

```python
path = '../../data/external/'
file = 'DT_parameters_2RC_5T_single_cell_C15.xlsx'

xls = pd.ExcelFile(path + file)
dfs = dfs = {sheet_name: xls.parse(sheet_name) 
          for sheet_name in xls.sheet_names}
```

```python
df = pd.concat([df for df in dfs.values()], ignore_index=True, axis=0)
```

```python
columns_to_drop = ['filename', 'data', 'start interval', 'end interval', 'skip', 'num value extrap', 'Rsqd', 'RMSE fitting']
df = df.drop(columns=columns_to_drop)
```
