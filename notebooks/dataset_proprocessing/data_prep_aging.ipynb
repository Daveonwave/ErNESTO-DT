### Creation of parameters

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

from src.postprocessing.visualization import ernesto_plotter
from src.postprocessing.metrics import compute_metrics, _mse
```

```python
pathfile = "../../data/external/paper_dic24/invecchiamento/Parametrizzazione_celle_modulo/DT_parameters_2RC_3T_4CU_module_avg_cell.xlsx"

xls = pd.ExcelFile(pathfile)
df_params = {sheet_name: xls.parse(sheet_name) for sheet_name in xls.sheet_names}
```

```python
sheets_to_read = ['CU1', 'CU2', 'CU3', 'CU4',]
```

```python
df_params[sheets_to_read[0]]
```

```python
folder = "../../data/external/paper_dic24/invecchiamento/Capacita_persa_invecchiamento/"
files = os.listdir(folder)

temps = ['10degC', '25degC', '40degC']
soh_dict = {}

for f in files:
    xls = pd.ExcelFile(folder + f)
    logic = [temp in f for temp in temps]
    soh_dict[np.array(temps)[logic][0]] = xls.parse('SoH')
```

```python
soh_dict
```

```python
soh = []

for cu in range(4):
    soh.append([soh_dict[temp]['SoH'][cu] for temp in temps])
```

```python
soh
```

```python
for i, sheet in enumerate(sheets_to_read):
    soh_values = [soh[i][0]] * 11 + [soh[i][1]] * 11 + [soh[i][2]] * 11
    df_params[sheet]['soh'] = soh_values
```

```python
df_params['CU4']
```

```python
df_complete = df_params['CU1']
```

```python
for sheet in sheets_to_read[1:]:
    df_complete = pd.concat([df_complete, df_params[sheet]], ignore_index=True)
```
