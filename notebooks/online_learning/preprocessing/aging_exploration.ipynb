```python
import pandas as pd
import os
from src.online_learning.utils import utils
import matplotlib.pyplot as plt
```

```python
relative_path_ground = os.path.join('..', '..', 'data', 'ground', 'ground_aging.csv')
path_ground_aging = utils.get_absolute_path(relative_path_ground)
df_aging = pd.read_csv(path_ground_aging)
df_aging = df_aging.drop(columns=['Unnamed: 0'])
```

```python
df_aging.describe()
```

```python
plt.figure(figsize=(15, 7))
df_aging['Current [A]'].iloc[:12000].plot(kind='line')
plt.xlabel('Index')
plt.ylabel('Current [A]')
plt.title('Current over Time')
plt.show()
```

```python
plt.figure(figsize=(15, 7))
df_aging['Voltage [V]'].iloc[:12000].plot(kind='line')
plt.xlabel('Index')
plt.ylabel('Voltage [V]')
plt.title('Voltage over Time ')
plt.show()
```

```python
plt.figure(figsize=(15, 7))
df_aging['Temperature [degC]'].iloc[:3000].plot(kind='line')
plt.xlabel('Index')
plt.ylabel('Temperature [degC]')
plt.title('Temperature over Time ')
plt.show()
```
