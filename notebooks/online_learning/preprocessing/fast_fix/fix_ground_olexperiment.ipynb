```python
df['Power(W)'] = df['Voltage(V)'] * df['Current(A)']
```

```python
df['Power(W)'].iloc[0:5000].plot()
```

```python
n_times = 10
dfs = []
for i in range(n_times):
    df_shifted = df.copy()
    df_shifted['Time'] = pd.to_datetime(df_shifted['Time'])
    df_shifted['Time'] += pd.Timedelta(seconds=i * len(df))
    dfs.append(df_shifted)
df_repeated = pd.concat(dfs, ignore_index=True)   
```
