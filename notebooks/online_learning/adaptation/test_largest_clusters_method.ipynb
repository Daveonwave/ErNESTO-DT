```python
from src.online_learning.change_detection.similarity_based_clustering import (mountain_method,
                                                                              largest_cluster, largest_clusters)
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean1 = np.array([5, 10, 15])  
cov1 = np.array([[7, 1, 0],  
                [1, 17, 2],
                [0, 2, 12]])
n_samples1 = 30
```

```python
mean2 = np.array([35, 40, 45])  
cov2 = np.array([[7, 1, 0],  
                [1, 17, 2],
                [0, 2, 12]])
n_samples2 = 20
```

```python
data_multivariate1 = np.random.multivariate_normal(mean1, cov1, n_samples1)
data_multivariate2 = np.random.multivariate_normal(mean2, cov2, n_samples2)
```

```python
fig = plt.figure(figsize=(16, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(data_multivariate1[:, 0], data_multivariate1[:, 1], data_multivariate1[:, 2], c='yellow', marker='o', alpha=0.6)

ax.scatter(data_multivariate2[:, 0], data_multivariate2[:, 1], data_multivariate2[:, 2], c='purple', marker='^', alpha=0.6, label='Uniform Distribution')

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Trivariate Gaussian Cluster')
plt.show()
```

```python
data = np.vstack((data_multivariate1, data_multivariate2))
outliers = data.tolist()
```

```python
centers = mountain_method(outliers=outliers, epsilon=0.001, radius=1, p=1)
```

```python
print(centers)
```

## RMK: In this specific experiment the more far away from each other are the centroid of the two clusters the more likely is that the method gets the correct centers

```python
biggest_cluster = largest_cluster(centers=centers,outliers=outliers, epsilon=0.001)
```

```python
print(biggest_cluster)
```
