```python
import pandas as pd
```

```python
df = pd.read_csv('saved_results/history_theta.csv')
```

```python
df.describe()
```

```python
df.drop(df.columns[0], axis=1)
```

```python
theta_history = df.to_numpy()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

r0 = theta_history[:, 0]
rc = theta_history[:, 1]
c = theta_history[:, 2]

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the data points
ax.scatter(r0, rc, c)

# Set labels
ax.set_xlabel('r0 Label')
ax.set_ylabel('rc Label')
ax.set_zlabel('c Label')

# Show the plot
plt.show()
```
